```python
from datetime import datetime
```

```python
current_date = datetime.now()
df['age'] = df['Birth Date'].apply(lambda x: current_date.year - x.year - ((current_date.month, current_date.day) < (x.month, x.day)))
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
label_encoding = preprocessing.LabelEncoder()

df['Gender'] = label_encoding.fit_transform(df['Gender'])

df['Married'] = label_encoding.fit_transform(df['Married'])

df['Dependents'] = label_encoding.fit_transform(df['Dependents'])

df['Internet Service'] = label_encoding.fit_transform(df['Internet Service'])

df['Premium Tech Support'] = label_encoding.fit_transform(df['Premium Tech Support'])

df['Streaming Movies'] = label_encoding.fit_transform(df['Streaming Movies'])

df['Streaming Music'] = label_encoding.fit_transform(df['Streaming Music'])

df['Paperless Billing'] = label_encoding.fit_transform(df['Paperless Billing'])
```

```python
a = ['Month-to-Month', 'One Year', 'Two Year']

df['Contract'] = df['Contract'].astype(str)

encode1 = OrdinalEncoder(categories=[a], handle_unknown='use_encoded_value', unknown_value=-1)
df['Contract'] = encode1.fit_transform(df[['Contract']])
```

```python
enc = OneHotEncoder()
aenc1 = pd.DataFrame(enc.fit_transform(df[['Internet Type']]).toarray(),columns=enc.get_feature_names_out())
benc1 = pd.DataFrame(enc.fit_transform(df[['Payment Method']]).toarray(),columns=enc.get_feature_names_out())

df = df.reset_index()
df = df.reset_index()

df = pd.concat([df,aenc1], axis=1)
df = pd.concat([df,benc1], axis=1)
```

```python
df = df.drop(['level_0', 'index'], axis = 1)
df = df.drop('Country', axis = 1)
df = df.drop('Customer ID', axis = 1)
df = df.drop('City', axis = 1)
df = df.drop('Birth Date', axis = 1)
df = df.drop('Internet Type', axis = 1)
df = df.drop('Payment Method', axis = 1)
```

```python
df.info()
```

```python
from sklearn.preprocessing import RobustScaler
scaling = RobustScaler()
```

```python
df['Number of Dependents'] = scaling.fit_transform(df[['Number of Dependents']])
df['Number of Referrals'] = scaling.fit_transform(df[['Number of Referrals']])
df['Avg Monthly GB Download'] = scaling.fit_transform(df[['Avg Monthly GB Download']])
df['Total Refunds'] = scaling.fit_transform(df[['Total Refunds']])
df['Satisfaction Score'] = scaling.fit_transform(df[['Satisfaction Score']])
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

k_range = range(2, 11)

sse = [KMeans(n_clusters=k, random_state=42).fit(df).inertia_ for k in k_range]
silhouette_scores = [silhouette_score(df, KMeans(n_clusters=k, random_state=42).fit_predict(df)) for k in k_range]
```

```python
# Plot Elbow Method
plt.figure(figsize=(10, 6))
plt.plot(k_range, sse)
plt.xlabel('Number of Clusters')
plt.ylabel('SSE (Sum of Squared Errors)')
plt.title('Elbow Method')
plt.show()
```

```python
# Plot Silhouette Scores
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(df)
```

```python
df['Cluster'] = kmeans.labels_
```

```python
from sklearn.decomposition import PCA
import seaborn as sns
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df.drop('Cluster', axis=1))
df['PCA1'] = df_pca[:, 0]
df['PCA2'] = df_pca[:, 1]

plt.figure(figsize=(12, 8))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df, palette='cool', s=80, edgecolor='black', alpha=0.8 )
plt.title('Cluster Visualization (PCA Projection)', fontsize=16, fontweight='bold')
plt.legend(title='Cluster', title_fontsize=12, fontsize=10, loc='upper right')
plt.grid(linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Select all columns except the cluster column for comparison
columns_to_plot = [col for col in df.columns if col != "Cluster"]

# Check if there are columns to plot
if len(columns_to_plot) == 0:
    print("No columns to plot.")
else:
    # Plot comparisons for each column grouped by cluster
    plt.figure(figsize=(15, len(columns_to_plot) * 5))
    for i, col in enumerate(columns_to_plot, 1):
        plt.subplot(len(columns_to_plot), 1, i)
        if df[col].dtype in ['object', 'category']:
            # Grouped bar plot for categorical columns
            grouped_data = df.groupby(["Cluster", col]).size().reset_index(name="counts")
            sns.barplot(x="Cluster", y="counts", hue=col, data=grouped_data, palette="Set2")
            plt.ylabel("Count", fontsize=14)
        else:
            # Box plot for numeric columns
            sns.boxplot(x="Cluster", y=col, data=df, palette="Set2")
            plt.ylabel(col, fontsize=14)

        plt.title(f'Distribution of {col} Across Clusters', fontsize=16)
        plt.xlabel("Cluster", fontsize=14)
        plt.legend(title=col, fontsize=12, loc="upper right") if df[col].dtype in ['object', 'category'] else None

    plt.tight_layout()
    plt.show()
```
